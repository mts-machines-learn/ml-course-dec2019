# Библиотека matplotlib


`plt.scatter(x, y, params)` — рисует точки с координатами $(x, y)$

`plt.plot(x, y, params)` — рисует график по точкам с координатами $(x, y)$

Вспомогательные функции

`plt.figure(figsize=(x, y))` — задать размер графика по вертикали и горизонтали

`plt.show()` — вывести график

`plt.subplot(...)` — добавить подграфик

`plt.xlim(x_min, x_max)` — пределы графика по оси $x$

`plt.ylim(y_min, y_max)` — пределы графика по оси $y$

`plt.title(name)` — имя графика

`plt.xlabel(name)` — название оси $x$

`plt.ylabel(name)` — название оси $y$

`plt.grid()` — добавить сетку на график

`plt.savefig(filename)` — сохранить график в файл


Много примеров здесь: http://matplotlib.org/gallery.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

Построим ломаную: зададим координаты y, координаты x по умолчанию - последовательность 0,1,2,...

```python
plt.figure()
plt.plot([0, 3, 1, 2, -4])
plt.show()
```

Координаты y можно также передать, как аргумент plt.plot

```python
plt.figure()
plt.plot([0, 1, 2, 3, 4], [0, 3, 1, 2, -4])
plt.show()
```

`scatter`  - нарисует точки с координатами (x,y), не соединяя их линиями

```python
plt.figure()
plt.scatter([0, 1, 2, 3, 4], [0, 3, 1, 2, -5])
plt.show()
```

Если точек очень много, то кривая станет гладкой

```python
x = np.arange(0, 2 * np.pi, 0.1)

plt.figure()
plt.plot(x, np.sin(x))
plt.show()
```

Координаты могут выражаться функциями, тогда можно строить различные параметрические функции:

```python
t = np.linspace(0, 2 * np.pi, 100)

plt.figure()
plt.plot(2*np.cos(t)+np.cos(2*t), 2*np.sin(t)-np.sin(2*t))
plt.show()
```

Для простой регулировки цветов и типов линий после пары $x$ и $y$ координат вставляется форматная строка. Первая буква определяет цвет (`'r'` - красный, `'b'` - синий и т.д.), дальше задаётся тип линии (`'-'` - сплошная, `'--'` - пунктирная, `'-.'` - штрих-пунктирная и т.д.).

```python
x = np.arange(0, 4 * np.pi, 0.01)

plt.figure()
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b--')
plt.show()
```

Различные маркеры:

```python
x = np.arange(0, 10, 1)

plt.figure()
plt.plot(x, x**2, 'o')
plt.plot(x, -x, '^')
plt.show()
```

Вот пример настройки почти всего, что можно настроить. Можно задать последовательность засечек на оси $x$ (и $y$) и подписи к ним (в них, как и в других текстах, можно использовать $\LaTeX$-овские обозначения). Задать подписи осей $x$ и $y$ и заголовок графика. Во всех текстовых элементах можно задать размер шрифта. Можно задать толщину линий и штрихи (так, на графике косинуса рисуется штрих длины 8, потом участок длины 4 не рисуется, потом участок длины 2 рисуется, потом участок длины 4 опять не рисуется, и так по циклу; поскольку толщина линии равна 2, эти короткие штрихи длины 2 фактически выглядят как точки). Можно задать подписи к кривым (legend); где разместить эти подписи тоже можно регулировать.

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), linewidth=2, color='b', dashes=[8, 4], label=r'$\sin x$')
plt.plot(x, np.cos(x), linewidth=2, color='r', dashes=[8, 4, 2, 4], label=r'$\cos x$')
plt.axis([0, 2 * np.pi, -1, 1])
plt.xticks(np.linspace(0, 2 * np.pi, 9),
           ('0',r'$\frac{1}{4}\pi$',r'$\frac{1}{2}\pi$',
            r'$\frac{3}{4}\pi$',r'$\pi$',r'$\frac{5}{4}\pi$',
            r'$\frac{3}{2}\pi$',r'$\frac{7}{4}\pi$',r'$2\pi$'),
           fontsize=20)
plt.yticks(fontsize=12)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$y$', fontsize=20)
plt.title(r'$\sin x$, $\cos x$', fontsize=20)
plt.legend(fontsize=20, loc=1)
plt.show()
```

## Поверхности

Все поверхности параметрические: $x=x(u,v)$, $y=y(u,v)$, $z=z(u,v)$. Если мы хотим построить явную поверхность $z=z(x,y)$, то удобно создать массивы $x=u$ и $y=v$ функцией `meshgrid`.

```python
fig = plt.figure(figsize=(20, 8))

#===============
#  First subplot
#===============
# set up the axes for the first plot
ax = fig.add_subplot(1, 2, 1, projection='3d')

# plot a 3D surface like in the example mplot3d/surface3d_demo
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='gnuplot',
                      antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5)

#===============
# Second subplot
#===============
# set up the axes for the second plot
ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y = np.mgrid[0:6*np.pi:0.25, 0:4*np.pi:0.25]
Z = np.sqrt(np.abs(np.cos(X) + np.cos(Y)))

surf = ax.plot_surface(X, Y, Z, cmap='autumn', cstride=2, rstride=2)
ax.set_xlabel("X-Label")
ax.set_ylabel("Y-Label")
ax.set_zlabel("Z-Label")
ax.set_zlim(0, 2)

plt.show()
```
