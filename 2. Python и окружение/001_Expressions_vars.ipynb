# Python cheat sheet

### Объекты, выражения и переменные

**Объект** — любой кусок данных в памяти: число, строка, список, словарь и т. д. В питоне всё — объект.

**Выражение** — любой код, в результате работы которого появляется объект. Почти любой кусок кода в питоне — это выражение.

Самое простое выражение — это просто написать единичку:

```python
1
```

Единичка — это валидный кусок кода на питоне и самое короткое выражение, которое можно придумать. Когда Питон выполнит это выражение, он создат в памяти объект, внутри которого будет лежать единичка.

Вот, например, строка:

```python
'foo'
```

Это тоже выражение. Когда Питон его выполняет, он создаёт в памяти объект, в котором лежит строка `'foo'`.

#### Переменные

Объекты создаются в памяти в рандомных местах, и мы не сможем к ним обратиться, если не положим объект в переменную.

**Переменная** — это просто имя, которое мы связаваем с объектом. Ещё говорят, что переменная *указывает* на объект. Любая переменная в питоне может указывать на любой объект. Когда мы используем переменную в коде, Питон просто подставляет вместо неё объект, на который она указывает.

Таким образом, большинство выражений в питоне **создают новые объекты** в памяти, а переменные (это тоже маленькие выражения) — **достают ранее созданный объект** из памяти.

Мы помещаем объет в переменную с помощью одинарного равно `=` (оператор присваивания).

**Слева** от оператора присваивания всегда стоит то, **куда** мы кладём объект.

**Справа** от оператора присваиваивания всегда стоит то, **откуда** мы берём объект.

Создадим две перемнные:

```python
age = 18
destination = 'London'
```

Мы можем придумывать переменным любые имена. Обычно, если в названии переменных есть несколько слов, их отделяют подчёриванием:

```python
number_of_cows = 42
```

Мы можем вывести содержимое переменных в консоль с помощью встроенной функции `print()`. Она может принимать на вход как одну, так и несколько переменных:

```python
print(age)
print(destination, number_of_cows)
```

Что произошло в этом примере:

1. Когда мы написали `age` *без оператора присваивания `=`*, питон увидел, что раньше мы объявили переменную с таким именем, и она указывает на некий объект.
2. Питон достал объект, на который указывает переменная `age`, и держит его в руках.
3. Питон видит, что мы использовали переменную `age` в качестве аргумента функции. Питон запускает функцию `print()` и передаёт полученный из `age` объект.

---
Мы можем положить в переменную любой другой объект в любое время.
В этом примере сначала положим в переменную число, а потом строку.

```python
age = 18
print(age)

age = 'eighteen'
print(age)
```

Как работает этот пример? Сначала мы создали объект с числом `18` и положили этот объект в переменную `age`. Потом мы создали новый объект — строку `'eighteen'` и положили его в ту же самую переменную. После этого, переменная указывает на объект `'eighteen'`, а не на `18`. **Переменная всегда указывает только на один объект в памяти**.

---
Что будет, если мы напишем:

```python
age = 'eighteen'
other_age = age

print(age, other_age)
print(age is other_age)
```

В этом примере мы создали объект `18` и положили его в переменную `age`. После этого мы получили объект из переменной `age` и положили его ещё и в переменную `other_age`. **Теперь две переменные указывают на один и тот же объект в памяти**. Это нормально. Сколько угодно переменных могут указывать на один и тот же объект. Когда мы получаем объект из переменной, она **не перестаёт** на него указывать.

#### Вложенные выражения

Выражения в питоне могут быть сколь угодно вложенными. Питон сначала выполняет самое вложенное выражение, получает от него объект, передаёт во внешнее выражение, получает от него новый объект, передаёт в ещё более внешнее, и так далее.

Разберём пример с функцией `len`

```python
len('foo')
```

![expressions](img/expressions.png)

Сначала мы создали в памяти новый объект — строку. Питон держит его в руках на тот случай, если мы захотим положить его в переменную или передать в функцию. Он видит, что вокруг выражения `'foo'` действительно написан вызов функции. 

Питон запускает функцию и передаёт в неё объект `'foo'`, который всё ещё держит в руках.

Функция обработала объект `'foo'` и создала нам новый объект — `3`. Теперь Питон «держит в руках» объект `3`. Так как мы этот объект не положили в переменную и не передали в другую функцию, мы его тут же потеряем.

Попробуем теперь не потерять объект `3`, а сразу передать его в другую функцию — `print`.

```python
print(len('foo'))
```

Что тут произошло? Три вложенных выражения. Объект `3` , который нам создала ф-ция `len` , мы тут же передали в функцию `print`.

### Сравнение объектов через `is`

Чтобы узнать, указывают ли две разные переменные на один и тот же объект, можно применить оператор `is`. Мы уже видели его в примере, теперь рассмотрим его более подробно.

```python
age = 'eighteen'
other_age = age
hello = 'Hello world'

print(age, other_age, hello)
print(age is other_age)
print(age is hello)
```

В этом примере мы создали объект со строкой `eighteen` и положили его в переменную `age`. После этого мы получили объект из переменной `age` и положили его ещё и в переменную `other_age`. **Теперь две переменные смотрят на один и тот же объект в памяти**.

Для разнообразия мы создали другую строку — `Hello world` — и положили её в третью переменную.

Оператор `is` возвращает `True` если слева и справа от него находится **один и тот же** объект. 

Здесь нужно понимать разницу: в памяти могут находиться два объекта с одинаковым содержимым, но это будут всё равно два **отдельных, разных** объекта. Мы можем поменять содержимое одного из них, и содержимое другого не поменяется. Для таких объектов оператор `is` вернёт `False`.

Если же у нас есть **один и тот же объект**, но на него просто указывают разные переменные, оператор `is` вернёт нам `True`.
